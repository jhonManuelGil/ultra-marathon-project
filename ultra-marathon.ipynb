### Sobre o conjunto de dados do projeto

- De acordo com a Wikipedia, uma ultramaratona, também chamada de ultra distância ou ultra corrida, é qualquer corrida a pé mais longa do que o comprimento tradicional da maratona de 42,195 quilômetros (26 mi 385 jardas). Várias distâncias são disputadas competitivamente, desde a ultramaratona comum mais curta de 31 milhas (50 km) até mais de 200 milhas (320 km). 50k e 100k são distâncias recordes do Atletismo Mundial, mas algumas corridas de 100 milhas (160 km) estão entre os eventos mais antigos e prestigiados, especialmente na América do Norte.

Os dados neste arquivo são uma grande coleção de registros de corridas de ultramaratona registrados entre 1798 e 2022 (um período de mais de dois séculos) sendo, portanto, uma amostra formidável de longo prazo. Todos os dados foram obtidos em sites públicos.


### O conjunto de dados contém 7.461.226 recordes de corridas de ultramaratona de 1.641.168 atletas únicos.
As seguintes colunas (com tipos de dados) são incluídas:

- Ano do evento (int64)
- Datas do evento (objeto)
- Nome do evento (objeto)
- Distância/comprimento do evento (objeto)
- Número de finalistas do evento (int64)
- Desempenho do atleta (objeto)
- Clube de atletas (objeto)
- País do atleta (objeto)
- Ano de nascimento do atleta (float64)
- Gênero do atleta (objeto)
- Categoria de idade do atleta (objeto)
- Velocidade média do atleta (objeto)
- ID do atleta (int64)

A coluna Distância/comprimento do evento descreve o tipo de corrida, cobrindo as distâncias e comprimentos de corrida mais populares da UM e algumas outras modalidades específicas (vários dias, etc.):

- Distâncias: 50km, 100km, 50mi, 100mi
- Comprimentos: 6h, 12h, 24h, 48h, 72h, 6d, 10d
- Além disso, há informações de idade, sexo e velocidade (em km/h) em outras colunas
