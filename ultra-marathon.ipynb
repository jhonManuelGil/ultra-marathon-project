```python

df2 = df[(df['Event distance/length'].isin(['50Km', '50mi'])) & (df['Year of event'] == 2020) &  (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')] 
# Store the filtered dataframe in df2

```

```python
df2.shape # Check the number of rows and columns in the filtered dataframe
```

```python
# remove (USA) from event name
df2['Event name'].str.split('(').str.get(0) # Update the 'Event name' column in df2
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0) # Update the 'Event name' column in df2
```

```python
df2.head() # Display the first 5 rows of the filtered dataset
```

```python
# clear up atlast  age column
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove  h from athlete performance 
df2['Athlete performance'] = df2['Athlete performance'].str.replace(r'h$', '', regex=True).str.strip() 
# Remove 'h' at the end and strip whitespace

```

```python
df2.head(5) # Display the first 5 rows of the cleaned dataset
```

```python
# drop unneeded columns Athlete club Athlete year of birth, Athlete age category
df2 = df2.drop(columns=['Athlete club', 'Athlete year of birth', 'Athlete age category', 'Athlete country'], axis=1)
```

```python
df2.head(5) # Display the first 5 rows of the cleaned dataset
```

```python
# remove  up null values
df2.isna().sum() # Check for null values in each column
```

```python
df2[df2['Athlete age'].isna() == 1] # Show rows with null values in 'Athlete age' column
```

```python
df2 = df2.dropna() # Drop rows with null values
```

```python
df2.shape # Check the number of rows and columns in the cleaned dataframe
```

```python
# chaeck for dupes
df2[df2.duplicated() == True] # Show duplicated rows
```

```python
# reset index
df2.reset_index(drop=True) # Reset the index of the cleaned dataframe
```

```python
# fix types
df2.dtypes # Check the data types of each column
```

```python
df2['Athlete age'] = df2['Athlete age'] .astype(int) # Convert 'Athlete age' column to integer type
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'] .astype(float) # Convert 'Athlete average speed' column to float type 
```

### Renome colums
- Year of event
- Event dates
- Event name
- Event distance/length
- Event number of finishers
- Athlete performance
- Athlete gender
- Athlete average speed
- Athlete ID
- Athlete age

```python
df2 = df2.rename(columns={
    'Year': 'year',
    'Event dates': 'event_dates',
    'Event name': 'event_name',    
    'Event distance/length': 'event_distance_length', 
    'Event number of finishers': 'event_number_of_finishers',  
    'Athlete performance': 'athlete_performance', 
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete name': 'athlete_name',
    'Athlete age': 'athlete_age',
    'Athlete ID': 'athlete_ID'    
    
 })
```

```python
df2.head()
```

```python
df2.columns.tolist()
# reorder columns
```

```python
df3 = df3 = df2[[
   'event_name',
    'event_dates',      
    'event_number_of_finishers',
    'event_distance_length',
    'athlete_average_speed', 
    'athlete_performance',  
    'athlete_ID',       
    'athlete_gender'
    
]] # Reorder columns in df2 and store in df3
```

```python
df3.head()
```

```python
df3 = df3.copy()
df3['event_name'] = df3['event_name'].str.strip()
df3['event_name'] = df3['event_name'].str.replace(r"\s+", " ", regex=True)
```

```python
# find 2 races that i ran in 2020 - sarasota / Evenglades 
df3[df3['event_name'] == 'Everglades 50 Mile Ultra Run'] 
```

```python
df3['athlete_ID'].value_counts() # Count occurrences of each athlete ID
```

```python
import matplotlib.pyplot as plt
```

```python
sns.histplot(df3['athlete_ID'], bins=30, kde=True) # Show distribution of athlete average speed
```

```python
sns.histplot(df3, x = 'athlete_performance', hue='athlete_gender') # Show distribution of athlete average speed
```

```python
sns.histplot(df3, x = 'athlete_average_speed', hue='athlete_gender') # Show distribution of athlete average speed
```

```python
df3['athlete_average_speed'].hist(bins=30)
plt.xlabel("Athlete average speed")
plt.ylabel("Frequency")
plt.title("Distribution of Athlete Average Speed")
plt.show()
```

```python
palette = {'M': 'green', 'F': 'purple'}

plt.figure(figsize=(10, 5)) # Set figure size

sns.boxplot(
    data=df3,
    x='athlete_average_speed',
    y='athlete_gender',
    hue='athlete_gender',  # usar hue
    palette=palette,
    dodge=False,            # para que no duplique cajas
    legend=False
)

plt.title("Speed Distribution by Gender") # Set plot title
plt.xlabel("Average Speed") # Set x-axis label
plt.ylabel("Gender") # Set y-axis label
plt.show()
```
